```python
from pymongo import MongoClient
```

```python
client = MongoClient("")
```

```python
db=client["test_ecom_db"]

collection=db["orders"]

print("connected to mongo db")
```

```python
order = {
    "name": "Men Casual Shirt",
    "price": 999,
    "category": "men"
}

collection.insert_one(order)
```

```python
products = [
    {"name": "Jeans", "price": 1499, "category": "men"},
    {"name": "Kurti", "price": 999, "category": "women"},
    {"name": "Kids Jacket", "price": 1299, "category": "kids"}
]

collection.insert_many(products)

print("Multiple Products Inserted Successfully")
```

```python
all_products = collection.find()

for p in all_products:
    print(p)
```

TEXT 2 Nosql

```python
product = collection.find_one({"name": "Kurti"})

print(product)
```

```python
men_products = collection.find({"category": "kids"})

for p in men_products:
    print(p)
```

UPDATE

```python
collection.update_one(
    {"name": "Jeans"},
    {"$set": {"price": 1399}}
)

print("Product Updated ")
```

DELETE

```python
collection.delete_one({"name": "Kids Jacket"})

print("Product Deleted")
```

```python
collection.delete_many({})

print("All Products Deleted Successfully")
```

DROP

```python
collection.drop()

print("Collection Dropped")
```
